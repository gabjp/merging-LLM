```python
def read_txt(path):
    file1 = open(path, 'r')
    Lines = file1.readlines()
    sums = []
    means = []
    colors = []

    for line in Lines:
        if "Sum: " in line:
            sums.append(float(line.split()[1]))
        elif "Mean: " in line:
            means.append(float(line.split()[1]))
        elif "mlp" in line:
            colors.append("r")
        elif "self_attn" in line:
            colors.append("b")
    
    assert len(means) == len(sums) == len(colors)

    return sums, means
```

```python
import matplotlib.pyplot as plt

def plot_histogram(numbers, xlabel, ylabel, title):
    plt.figure(figsize=(10, 8))
    plt.bar(range(len(numbers)), numbers)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()

```

```python
sums, means = read_txt('task_vectors_dist/sciq.txt')
plot_histogram(sums, 'layer', 'sum', 'sum(abs(sciq - llama)) for each layer')
```

```python
plot_histogram(means, 'layer', 'mean', 'mean(abs(sciq - llama)) for each layer')
```
