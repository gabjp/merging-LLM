```python
def plot(categories, data, title):
    # Number of categories
    num_categories = len(categories)

    # Create a list of angles for the radar chart
    angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()

    # Create a radar chart
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)  # Create a polar plot

    # Plot the data
    data += data[:1]  # Close the loop by adding the first data point to the end
    angles += angles[:1]  # Close the loop by adding the first angle to the end
    ax.fill(angles, data, 'b', alpha=0.1)  # Fill the area under the data lines
    ax.plot(angles, data, 'b', linewidth=2)  # Plot the data lines

    # Set the labels for each category
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    ax.set_rticks([0.5*i for i in range(1,11)])

    # Add a title
    plt.title(title)

    # Show the radar chart
    plt.show()
```

```python
def multiplot(categories, datas, labels, title):
    # Number of categories
    num_categories = len(categories)

    # Create a list of angles for the radar chart
    angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()
    angles += angles[:1]

    # Create a radar chart
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)  # Create a polar plot

    # Plot the data
    for data, label in zip(datas, labels):
        data += data[:1]  # Close the loop by adding the first data point to the end
        
        ax.plot(angles, data, linewidth=2, label = label)  # Plot the data lines

    # Set the labels for each category
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    ax.set_rticks([0.5*i for i in range(1,11)])

    # Add a title
    plt.title(title)
    plt.legend(loc = (1,1))
    # Show the radar chart
    plt.show()
```
